```
#changing data type of column
df = df.withColumn("gender", df.gender.cast("Integer"))
```

```
df.printSchema()
```

```
display(df)
```

```
#updating values of column
df = df.withColumn("id", df.id + 100)
```

```
display(df)
```

```
#using lit function to assign defalt value to all items in the column
from pyspark.sql.functions import *
df = df.withColumn("gender", lit('Female'))
```

```
display(df)
```

```
#dropping a column
df = df.drop("ip_address")
```

```
display(df)
```

```
#renaming the column name
df = df.withColumnRenamed("first_name", "first Name")
```

```
display(df)
```

```
#adding new column based on existing column
df = df.withColumn("id2", df.id)
```

```
display(df)
```

```
#dropping null values
df.na.drop()
```

```
display(df)
```

```
#using filter function to filter data
display(df.filter(df.last_name == "Flament"))
display(df.filter(df.id > 20))
```
